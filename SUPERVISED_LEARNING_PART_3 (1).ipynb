KNN - K NEAREST NEIGHBORS

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn. neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

DATA SET LOADING

```python
Dataset=load_iris()

print(Dataset)
```

```python
x=Dataset.data
y=Dataset.target
```

```python
import pandas as pd
df=pd.DataFrame(x,columns=Dataset.feature_names)
print(df)
```

TARGET

```python
print(y)
```

```python
print(set(y))
Dataset.target_names
```

```python
df["target"]=y
print(df)
```

TEST TRAIN SPLIT

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train)
```

SCALER

```python
std_scaler=StandardScaler()
mm_scaler=MinMaxScaler()
x_train_std=std_scaler.fit_transform(x_train)
x_train_mm=mm_scaler.fit_transform(x_train)
print(x_train_std)

```

```python
print(x_train_mm)
```

```python
x_test=mm_scaler.transform(x_test)
print(x_test)
```

```python
Flowers = KNeighborsClassifier(n_neighbors= 5)
Flowers.fit(x_train_mm,y_train)
```

```python
y_pred=Flowers.predict(x_test)
```

```python
print("ACCURACY=",accuracy_score(y_test,y_pred))
print("\n\nCLASSIFICATION:\n\n",classification_report(y_test,y_pred))
print("\n\nCONFUSION MATRIX:\n\n",confusion_matrix(y_test,y_pred))
```

```python
print("ACCURACY = ",accuracy_score(y_test,y_pred))
print("\n\nCLASSIFICATION REPORT :\n\n",classification_report(y_test,y_pred))
print("\n\nCONFUSION MATRIX :\n\n",confusion_matrix(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CFM= confusion_matrix(y_test,y_pred)
labels=Dataset.target_names


plt.figure(figsize=(6,4))

sns.heatmap(CFM,annot=True,fmt="d",xticklabels=labels,yticklabels=labels,cmap="Blues")

plt.title("CONFUSION MATRIX - HEATMAP")
plt.xlabel("PREDICTED CONDITION")
plt.ylabel("ACTUAL CONDITION")
plt.show()
```
